```python
from keras.layers.core import Dense, Activation, Dropout
from keras.layers.recurrent import LSTM
from keras.models import Sequential
import lstm, time #helper libraries
import matplotlib.pyplot as plt
```

```python
seq_len=50

#Step 1 Load Data
X_train, y_train, X_test, y_test = lstm.load_data('googl5.csv', seq_len, True)
```

```python
#Step 2 Build Model
model = Sequential()

model.add(LSTM(
    input_dim=1,
    output_dim=50,
    return_sequences=True))
model.add(Dropout(0.2))

model.add(LSTM(
    100,
    return_sequences=False))
model.add(Dropout(0.5))

model.add(Dense(
    output_dim=1))
model.add(Activation('linear'))

start = time.time()
model.compile(loss='mse', optimizer='rmsprop')
print('compilation time : ', time.time() - start)
```

```python
#Step 3 Train the model
model.fit(X_train, y_train, batch_size=512, nb_epoch=500, validation_split=0.05)
```

```python
#Step 4 - Plot the predictions!

predictions = lstm.predict_sequences_multiple(model, X_test, seq_len, seq_len)
lstm.plot_results_multiple(predictions, y_test, seq_len)

#plt.plot(predictions, color='red', label='prediction')
#plt.plot(y_test,color='blue', label='y_test')
#plt.legend(loc='upper left')
#plt.show()
```

```python
import numpy as np
predictions = np.array(predictions)
predictions
```

```python
# Con esto hayamos cuales de las predicciones han acertado la tendencia.
diff = [(x/abs(x)) for x in predictions[0,:-1]-predictions[0,1:]]
y_test = np.array(y_test, dtype=float)
diff_real = [(x/abs(x)) for x in y_test[:-1]-y_test[1:]]
res = []
for x in range(len(diff)):
    res.append(diff_real[x] == diff[x])
res.sort()
res
```
